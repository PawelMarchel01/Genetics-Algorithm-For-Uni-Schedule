```python
class Weekdays:
    days=["Poniedziałek", 'Wtorek','Środa','Czwartek','Piątek']
```

```python
class Professor:

    #initiating Proffesor
    def __init__(self, name):
        self.name=name
        self.class_assignment = pd.DataFrame(columns=Weekdays.days, index=['8-10','10-12','12-14','14-16','16-18'])

    def assign_classes(self, lecture):
        day = lecture.day
        time = lecture.time
        if pd.isnull(self.class_assignment.iloc[time, day])==False:
            raise(ValueError('Trying to assign a class to ocupied lecturer!'))
        else:
            self.class_assignment.iloc[time, day]=lecture
```

```python
class Subject:

    #Seting room category
    other_classes = 0
    labolatories = 1
    pe = 2
    english = 3

    #initiating Subject
    def __init__(self, name, lecturer, students, room = 'brak' ,classtype='other_classes'):
        self.name=name
        self.lecturer=lecturer
        self.students=students
        if classtype == 'other_classes':
             self.classtype = 0
        elif classtype == 'pe':
             self.classtype = 2
        elif classtype == 'english':
             self.classtype = 3
        else:
             self.classtype = 1
        self.room='brak'

    def add_room(self, name):
        self.room = name

    def __str__(self):
        return self.name + '\t Prowadzący: ' + self.lecturer.name + '\t Sala: '  + self.room
```

```python
class Students:

    #initiating Students
    def __init__(self, year):
        self.year = year
        self.class_assignment = pd.DataFrame(columns=Weekdays.days, index=['8-10','10-12','12-14','14-16','16-18'])

    def assign_classes(self, lecture):
        day = lecture.day
        time = lecture.time
        if pd.isnull(self.class_assignment.iloc[time, day])==False:
            raise(ValueError('Trying to assign a class to ocupied students!'))
        else:
            self.class_assignment.iloc[time, day]=lecture
```

```python
class Room:

    #seting room category
    standard = 0
    labolatories = 1
    pe = 2
    english = 3

    #initiating Room
    def __init__(self, name, category='standard'):
        self.name = name
        if category == 'standard':
             self.category = Room.standard
        elif category == 'pe':
             self.category = Room.pe
        elif category == 'english':
             self.category = Room.english
        else:
             self.category = Room.labolatories
        self.class_assignment = pd.DataFrame(columns=Weekdays.days, index=['8-10','10-12','12-14','14-16','16-18'])

    def assign_classes(self, lecture):
        day = lecture.day
        time = lecture.time
        if pd.isnull(self.class_assignment.iloc[time, day])==False:
            raise(ValueError('Trying to assign a class to ocupied room!'))
        else:
            self.class_assignment.iloc[time, day]=lecture
```

```python
class University:

    #initiating University
    def __init__(self):
        self.professors={}
        self.grades={}
        self.rooms={}

    #addin room
    def add_room(self, name, classtype='standard'):
        self.rooms[name]=Room(name, classtype)

    #addin grade
    def add_grade(self, name, year):
        self.grades[name]=Students(year)

    #addin professor
    def add_professor(self, name):
        self.professors[name]=Professor(name)
```

```python
class fitness_class:

  import random
  from numpy import random as rand
  import copy

  #Fitness do badania powtórek. Pętla która sprawdza czy w tym samym dniu i o tej samej godzinie prowadzący lub rok nie ma takich samych zajęć <- angielski
  def fitness(x):

    score = 0

    for category in range(len(x)):
      for room in range(len(x[category])):
        for subject in range(len(x[category][room])):
          for next_category in range(len(x)):
            for next_room in range(len(x[next_category])):
              if next_category == category and next_room == room:
                score = score
              else:
                #print('powtórka kategoria ' + str(category) + 'sala ' + str(room) +  'przedmiot ' + str(subject) +'next kategoria: ' + str(next_category) + 'next sala: ' + str(next_room))
                if type(x[category][room][subject]) == Subject and type(x[next_category][next_room][subject]) == Subject and x[category][room][subject].lecturer.name == x[next_category][next_room][subject].lecturer.name:
                  #print('kategoria ' + str(category) + 'sala ' + str(room) +  'przedmiot ' + str(subject) +'next kategoria: ' + str(next_category) + 'next sala: ' + str(next_room))
                  score = score + 1
                if type(x[category][room][subject]) == Subject and type(x[next_category][next_room][subject]) == Subject and x[category][room][subject].students.year == x[next_category][next_room][subject].students.year:
                  score = score + 1
                  #print('kategoria ' + str(category) + 'sala ' + str(room) +  'przedmiot ' + str(subject) +'next kategoria: ' + str(next_category) + 'next sala: ' + str(next_room))

    return(score)

  def fitness_2(x):

    score = 0

    for category in range(len(x)):
      for room in range(len(x[category])):
          #print("kategoria " + str(category) + "sala " + str(room))
        if type(x[category][room][0]) == float: #poniedziałek 8
          score = score - 2
            #print("poniedzialek")
        if type(x[category][room][1]) == float: #poniedziałek 10
          score = score - 1
            #print("poniedzialek")
        if type(x[category][room][24]) == float: #piątek 20
            #print("piatek")
          score = score - 3
        if type(x[category][room][23]) == float: #piątek 20
            #print("piatek")
          score = score - 2
        if type(x[category][room][22]) == float: #piątek 20
            #print("piatek")
          score = score - 1
        if type(x[category][room][21]) == float: #piątek 20
            #print("piatek")
          score = score - 1
        if type(x[category][room][20]) == float: #piątek 20
            #print("piatek")
          score = score - 1

    return(score)

  def fitness_3(x):

    score = 0

    for category in range(len(x)):
      for room in range(len(x[category])):
        for subject in range(len(x[category][room])-1):
          for next_category in range(len(x)):
            for next_room in range(len(x[next_category])):
              if next_category == category and next_room == room:
                score = score
              else:
                if type(x[category][room][subject]) == Subject and type(x[next_category][next_room][subject+1]) == Subject and x[category][room][subject].lecturer.name == x[next_category][next_room][subject+1].lecturer.name:
                  score = score - 2
                if type(x[category][room][subject]) == Subject and type(x[next_category][next_room][subject+1]) == Subject and x[category][room][subject].students.year == x[next_category][next_room][subject+1].students.year:
                  score = score - 2

    return score

```

```python
import copy

class genetic_algorithm:

  def __init__(self, number_pop, dane):
    self.number_pop = number_pop
    self.dane = dane


  def population_generation(self):

    population = []

    for i in range(self.number_pop):
      dane_2 = copy.deepcopy(self.dane)
      self.plan = Initializing(dane_2, 'Politechnika_Lodzka')
      population.append(self.plan.plan_to_list())

    self.population = population

  def crossover_operator(self):

    for index in range(self.number_pop):

      dziecko = copy.deepcopy(self.population[index])

      for i in range(len(self.population[index])):

          mutation_prob = random.uniform(0, 1)
          s = random.randint(1,24) #Każda sala ma 25 pozycji więc bierzemy od 1 do 24 żeby działały permutacje
          x = random.sample(range(0, len(self.population[index][i])), 2) #Wybór 2 losowych sal

          #KRZYŻOWANIE
          mama = self.population[index][i][x[0]].copy()
          tata = self.population[index][i][x[1]].copy()
          mama_primprim = mama.copy()

          for j in range(s):
            mama[j] = tata[j]
            tata[j] = mama_primprim[j]

          if (mutation_prob > 0.5):
            #FUNCKJA 2 PERMUTACJE
            mama_prim = mama.copy()
            tata_prim = tata.copy()

            element = random.sample(range(0, 25), 2) #Wybór 2 losowych sal
            mama[element[0]] = mama[element[1]]
            mama[element[1]] = mama_prim[element[0]]

            #FUNCKJA 3 PERMUTACJE
            element = random.sample(range(0, 25), 2) #Wybór 2 losowych sal
            tata[element[0]] = tata[element[1]]
            tata[element[1]] = tata_prim[element[0]]

          dziecko[i][x[0]] = mama
          dziecko[i][x[1]] = tata

      self.population.append(dziecko)

  def cut_fitness(self):

    del_index = []

    for i in range(len(self.population)):
      if fitness_class.fitness(self.population[i]) != 0:
        del_index.append(i)

    del_index.sort(reverse=True)

    for index in del_index:
        if 0 <= index < len(self.population):
            self.population.pop(index)

  def select_by_fitness_2(self):

    calc_fitness_2 = []
    calc_fitness_3 = []
    for i in range(len(self.population)):
      calc_fitness_2.append(fitness_class.fitness_2(self.population[i]))
      calc_fitness_3.append(fitness_class.fitness_3(self.population[i]))

    for j in range(len(calc_fitness_2)):
      calc_fitness_2[j] = calc_fitness_2[j] + calc_fitness_3[j]

    combined_list = list(zip(self.population, calc_fitness_2))
    sorted_combined_list = sorted(combined_list, key=lambda x: x[1])
    self.population, calc_fitness_2 = zip(*sorted_combined_list)
    self.population = list(self.population[0:self.number_pop])
    self.calc_fitness_3 = list(calc_fitness_2[0:self.number_pop])
```

```python
class heuristic:

  def __init__(self, number_pop, dane, cycle_number):
    self.number_pop = number_pop
    self.dane = dane
    self.cycle_number = cycle_number

  def execute(self):

    self.ancestors = genetic_algorithm(number_pop = self.number_pop, dane = self.dane)

    self.ancestors.population_generation()

    for i in range(self.cycle_number):
        self.ancestors.crossover_operator()
        self.ancestors.cut_fitness()
        self.ancestors.select_by_fitness_2()
        if i == 1:
          print("First fitness score: " + str(self.ancestors.calc_fitness_3))
        if i == self.cycle_number-1:
          print("Last fitness score: " + str(self.ancestors.calc_fitness_3))

```

```python
class plan:

  def __init__(self, wynik_heurystyki):
    self.wynik_heurystyki = wynik_heurystyki
    self.best = self.wynik_heurystyki.ancestors.population[0]
    self.sale_name = list(self.wynik_heurystyki.ancestors.plan.university.rooms)

  def day_time_room(self):
    for category in range(len(self.best)):
      for room in range(len(self.best[category])):
        for i in range(len(self.best[category][room])):
          if type(self.best[category][room][i]) == Subject:
            self.best[category][room][i].day = i // 5
            self.best[category][room][i].time = i % 5
            self.best[category][room][i].room = self.sale_name[0]
        del self.sale_name[0]

  def empty_the_plan(self):
    for category in range(len(self.best)):
      for room in range(len(self.best[category])):
        for i in range(len(self.best[category][room])):
          if type(self.best[category][room][i]) == Subject:
            self.best[category][room][i].lecturer.class_assignment = self.best[category][room][i].lecturer.class_assignment.applymap(lambda x: np.NaN)
            self.best[category][room][i].students.class_assignment = self.best[category][room][i].students.class_assignment.applymap(lambda x: np.NaN)

  def clenerka(self):
    self.sale = list(self.wynik_heurystyki.ancestors.plan.university.rooms.values())
    for i in range(len(self.sale)):
      self.sale[i].class_assignment = self.sale[i].class_assignment.applymap(lambda x: np.NaN)

  def class_assigment(self):
    for category in range(len(self.best)):
      for room in range(len(self.best[category])):
        for i in range(len(self.best[category][room])):
          if type(self.best[category][room][i]) == Subject:
            for prof in range(len(self.wynik_heurystyki.ancestors.plan.teachers)):
              if self.best[category][room][i].lecturer.name == self.wynik_heurystyki.ancestors.plan.teachers[prof].name:
                self.wynik_heurystyki.ancestors.plan.teachers[prof].assign_classes(self.best[category][room][i])
            self.best[category][room][i].students.assign_classes(self.best[category][room][i])
            for j in range(len(self.sale)):
              if self.sale[j].name == self.best[category][room][i].room:
                self.sale[j].assign_classes(self.best[category][room][i])

  def list_to_df(self):
    self.day_time_room()
    self.empty_the_plan()
    self.clenerka()
    self.class_assigment()
```
